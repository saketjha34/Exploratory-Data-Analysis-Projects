```python
parent_edu_count  = sns.countplot(x = 'ParentEduc' , data = df)
plt.title('count of students vs parent education ')
plt.plot()
for bars in parent_edu_count.containers:
    parent_edu_count.bar_label(bars)
```

```python
def ap_five():
    bins =[]
    for i in range(0,100,5):
        bins.append(i)
    return bins 
     
sns.histplot(df['MathScore'] , bins=ap_five() , kde = True , color = c_green)
plt.plot()
```

```python
sns.histplot(df['ReadingScore'] , bins=ap_five() , kde = True , color = c_yellow)
plt.plot()
```

```python
sns.histplot(df['WritingScore'] , bins=ap_five() , kde = True , color = c_blue)
plt.plot()
```
